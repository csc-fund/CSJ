```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import csv 
import re
from py2neo import *
import random
import matplotlib.pyplot as plt
import matplotlib.dates as mdate
from datetime import datetime
```

```python
#导入文件
df = pd.read_csv('C:\\Users\\Administrator\\Desktop\\code_merged.csv', dtype=(object))
df = df.drop(columns='Unnamed: 0')
df
```

```python
##连接neo4j数据库，输入地址、用户名、密码
graph = Graph("http://localhost:7474", auth=("neo4j", "123456"))
```

```python
node_matcher = NodeMatcher(graph)                                      
relationship_matcher = RelationshipMatcher(graph)
number_news = {}
number_news_code = {}
for n in range(len(df.s_info_name)):
    a = node_matcher.match('stock', stock_code = df.s_info_code[n]).first()
    number_news[n] = a['过去一周新闻出现总次数']
number_news = pd.DataFrame([number_news])
number_news =  pd.DataFrame(number_news.values.T, index=number_news.columns, columns=number_news.index)
df = pd.concat([df,number_news],axis=1,ignore_index=True)
df.columns = ['code','name','indus_code','indus_name','mv','number']
df['mv']=df['mv'].astype('float')
df['number']=df['number'].astype('float')
df = df.fillna(0)
grouped=df.groupby(['indus_code'])
quantile_mv_75 = grouped['mv'].quantile(0.75)
quantile_mv_50 = grouped['mv'].quantile(0.5)
quantile_mv_25 = grouped['mv'].quantile(0.25)
quantile_news_75 = grouped['number'].quantile(0.75)
quantile_news_50 = grouped['number'].quantile(0.5)
quantile_news_25 = grouped['number'].quantile(0.25)
df = pd.merge(df,quantile_mv_75,on = ['indus_code'])
df = pd.merge(df,quantile_mv_50,on = ['indus_code'])
df = pd.merge(df,quantile_mv_25,on = ['indus_code'])
df = pd.merge(df,quantile_news_75,on = ['indus_code'])
df = pd.merge(df,quantile_news_50,on = ['indus_code'])
df = pd.merge(df,quantile_news_25,on = ['indus_code'])
df.columns = ['code','name','indus_code','indus_name','mv','number','mv_75','mv_50','mv_25','number_75','number_50','number_25']
df
```

```python
# create a list of our conditions
conditions_mv = [
    (df['mv'] <= df['mv_25'] ),
    (df['mv']  > df['mv_25']) & (df['mv']  <= df['mv_50']),
    (df['mv']  > df['mv_50']) & (df['mv']  <= df['mv_75']),
    (df['mv']  > df['mv_75'])
    ]
conditions_news = [
    (df['number'] <= df['number_25'] ),
    (df['number']  > df['number_25']) & (df['number']  <= df['number_50']),
    (df['number']  > df['number_50']) & (df['number']  <= df['number_75']),
    (df['number']  > df['number_75'])
    ]
values = ['1', '2', '3', '4']

df['mv_tier'] = np.select(conditions_mv, values)
df['news_tier'] = np.select(conditions_news, values)
df.head()
```
